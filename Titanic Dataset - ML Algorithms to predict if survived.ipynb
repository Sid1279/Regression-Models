```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import tkinter as tk
```

```python
titanic =  pd.read_csv("train.csv")
titanic
```

```python
#Make gender more readable (1 or 0) instead of M or F
titanic["Sex"] = titanic["Sex"].replace(["male", "female"], [1, 0])
titanic
```

```python
#filling NaN of age with mean value of age
titanic["Age"].fillna(titanic["Age"].mean(), inplace=True)
titanic
```

```python
#filling NaN of embarked with mode value of embarked
titanic["Embarked"].fillna(titanic["Embarked"].mode()[0], inplace=True)
titanic
```

```python
#Make Embarked more readable (2, 1 or 0) instead of S, C or Q
titanic["Embarked"] = titanic["Embarked"].replace(["S", "C", "Q"], [0, 1, 2])
titanic
```

```python
#We assume PassengerID, Pclass, Name, Ticket does not matter for survival
new_titanic = titanic.drop(["PassengerId", "Name", "Ticket", "Fare", "Cabin"], axis = 1)
new_titanic = new_titanic.drop(0)
new_titanic
```

```python
#Split into X and Y (for train-test-split)
y = new_titanic[["Survived"]]
X = new_titanic.drop("Survived", axis = 1)
```

```python
#80%-20% split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train.values, y_train)
```

```python
def will_survive(Pclass, Sex, Age, SibSp, Parch, Embarked):
    
    pronoun = "he" if Sex == 1 else "she"
    
    
    y_pred = clf.predict([[Pclass, Sex, Age, SibSp, Parch, Embarked]])
    prediction = y_pred[0]
    
    if prediction == 0:
        print("Sorry,", pronoun, "will not survive")
    else:
        print("Phew,", pronoun, "will survive")
```

```python
#User inputs
Pclass = input("Passenger Class (1/2/3): ")

Gender = input("Gender (M/F): ").lower()
Sex = 1 if Gender == "m" else 0

Age = input("Age: ")

SibSp = input("# of siblings/spouses aboard: ")

Parch = input("# of parents/children aboard: ")

Embarked_key = input("Embarked from Southampton(S) /Cherbourg (C) /Queenstown (Q)? : ").lower()
Embarked_map = {"s": 0, "c": 1, "q":2}
Embarked = Embarked_map[Embarked_key]


#Predict
will_survive(Pclass, Sex, Age, SibSp, Parch, Embarked)
```

Model Comparisons

```python
#80%-20% split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
def will_survive_raw(Pclass, Sex, Age, SibSp, Parch, Embarked):
    y_pred = clf.predict([[Pclass, Sex, Age, SibSp, Parch, Embarked]])
    prediction = y_pred[0]
    return prediction
```

```python
#Testing data - Prediction
rows = X_test.shape[0]
pred_arr = []
for i in range(rows):
    x = will_survive_raw(X_test.iloc[i]["Pclass"], X_test.iloc[i]["Sex"], X_test.iloc[i]["Age"], X_test.iloc[i]["SibSp"], X_test.iloc[i]["Parch"], X_test.iloc[i]["Embarked"])
    pred_arr.append(x)
    
#Testing data - Actual
actual_arr = []
for i in range(rows):
    x = y_test.iloc[i]["Survived"]
    actual_arr.append(x)


#measuring accuracy of our model
accurate = 0
for i in range(rows):
    if actual_arr[i] == pred_arr[i]:
        accurate +=1

accuracy = accurate/rows*100
print("The model is", accuracy,"% accurate")
```
